```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create two subplots with a shared y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

# Plot the data on each subplot
ax1.plot(x, y1)
ax1.set_title('Plot 1')
ax2.plot(x, y2)
ax2.set_title('Plot 2')

# Set the x-axis label for the entire figure
fig.text(0.5, 0.04, 'X-axis Label', ha='center')

# Set the y-axis label for the figure
fig.text(0.04, 0.5, 'Y-axis Label', va='center', rotation='vertical')

# Display the figure
plt.show()
```
