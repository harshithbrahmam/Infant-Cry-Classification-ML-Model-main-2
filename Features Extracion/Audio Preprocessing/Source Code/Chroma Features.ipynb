#### Description Of Piano
A piano is a keyboard instrument that produces sound by striking strings with hammers. The keys on a piano correspond to different notes, which are arranged in a repeating pattern called an octave.

In Western music, there are 12 different notes, which are named using the letters A, B, C, D, E, F, and G. These notes can be sharp (#) or flat (b), which means they are raised or lowered by a half-step, respectively.

The notes on a piano are arranged in a specific pattern of white and black keys. The white keys correspond to the natural notes (A, B, C, D, E, F, and G), while the black keys correspond to the sharp or flat notes (A#, C#, D#, F#, and G#). The black keys are arranged in groups of two and three between the white keys, forming a repeating pattern.

The piano keyboard spans over several octaves. An octave consists of 12 notes, where the 12th note is double the frequency of the 1st note. Each octave is divided into seven natural notes and five sharp or flat notes. The most common octave designation used in Western music is the scientific pitch notation, where each note is given a letter name followed by a number that indicates the octave.

For example, the note A4 represents the A note in the fourth octave, which is the A note that falls between the notes A5 and A3. The note C4 represents the C note in the fourth octave, which is the C note that falls between the notes C5 and C3.

In terms of half steps, each key on the piano keyboard is separated by a half step, except for the white keys that are adjacent to a black key, which are separated by a whole step. The distance between any two notes on the piano keyboard can be measured in terms of half steps. For example, the distance between C4 and C#4 (or Db4) is a half step.

In summary, the piano keyboard is arranged in a specific pattern of white and black keys that correspond to different notes in a repeating pattern of octaves. The scientific pitch notation is commonly used to identify the notes and octaves on a piano, with each note represented by a letter name followed by a number that indicates the octave. Each key on the piano keyboard is separated by a half step, except for the white keys that are adjacent to a black key, which are separated by a whole step.


#### Chroma Features
Chroma-based features, which are also referred to as "pitch class profiles", are a powerful tool for analyzing music whose pitches can be meaningfully categorized (often into twelve categories) and whose tuning approximates to the equal-tempered scale. 


Chroma features are a type of audio feature that represent the pitch class content of an audio signal, regardless of the octave in which it occurs. They are computed by mapping the frequency spectrum of an audio signal to a set of 12 pitch classes corresponding to the 12 notes in the Western musical scale (A, A#, B, C, C#, D, D#, E, F, F#, G, G#), and summing the energy of each pitch class across all frequencies within a given band.

There are several methods for computing chroma features, including:

Chroma vector method: This method involves dividing the audio spectrum into a set of frequency bands, typically using a logarithmic scale to better represent the way humans perceive pitch. Each frequency band is then mapped to one of 12 pitch classes, and the energy of each pitch class is computed by summing the energy of all frequencies within the corresponding band. The resulting chroma feature vector is typically normalized to have unit length.

Chroma STFT method: This method involves computing a short-time Fourier transform (STFT) of the audio signal, and then applying the chroma vector method to each time frame of the STFT. The resulting chroma feature matrix has dimensions of 12 x N, where N is the number of time frames in the STFT.

Chroma CQT method: This method involves computing a constant-Q transform (CQT) of the audio signal, which is a type of frequency-domain representation that is well-suited for musical analysis. The chroma vector method is then applied to each octave band of the CQT, resulting in a chroma feature matrix with dimensions of 12 x M, where M is the number of CQT bins.

Chroma features are commonly used in music information retrieval (MIR) tasks, such as music genre classification, chord recognition, and melody extraction. They can also be used as input features for machine learning models for tasks such as music recommendation or mood detection. The choice of method for computing chroma features depends on the specific task and the characteristics of the audio signal being analyzed.


In music, pitch refers to the perceived highness or lowness of a sound, and pitch classes are groups of pitches that share the same octave equivalence. For example, all the notes in the C major scale belong to the same pitch class, and they are separated by different octaves.

The chroma feature of a pitch class refers to the tonal quality or "color" of the sound. Chroma is a pitch class descriptor that is derived from the pitch class by ignoring the octave of the note. This means that notes with the same pitch class but different octaves will have the same chroma value.

For example, the notes C3, C4, and C5 all belong to the same pitch class (C) but different octaves. Their chroma values, however, are all the same because they share the same tonal quality or "color" of the sound.

Pitch height, on the other hand, refers to the specific octave of the note. This is important because notes in different octaves can have different timbres and harmonics that affect the overall sound quality.

By separating the pitch class into chroma and pitch height, we can analyze and compare the tonal qualities of different pitches without being influenced by their octave differences. This can be useful in music analysis, transcription, and synthesis.

#### Chroma Vector:
The Chroma vector is a feature vector that is used to represent musical pitches in a way that is more perceptually meaningful to humans. The concept of chroma refers to the tonal quality of a musical pitch, which is a property that is invariant to changes in octave. In other words, a C note played in one octave will have the same chroma as a C note played in a different octave.

The Chroma vector represents the magnitude of each of the 12 pitch classes in a standard chromatic scale. The chromatic scale includes all 12 notes in Western music, including both natural notes (A, B, C, D, E, F, G) and sharp or flat notes (A#, C#, D#, F#, and G#). Each pitch class in the Chroma vector corresponds to a specific note in the chromatic scale, and represents the amount of energy or activity in that note within a given musical passage or piece.

The Chroma vector is calculated using a method called cyclic helix representation, which maps the 12 pitch classes onto a circular or helical structure. This representation takes into account the periodicity of the pitch classes and ensures that the Chroma vector is invariant to changes in octave. This means that the Chroma vector will have the same values regardless of whether the musical passage is played in a low or high octave.

Overall, the Chroma vector is a useful feature for analyzing and representing musical pitch information, as it captures the perceptually meaningful tonal quality of pitches and is invariant to changes in octave. It is commonly used in music information retrieval, audio signal processing, and other fields that involve analyzing and synthesizing musical content.

#### Harmonic Content of Music
Chroma features are commonly used in music information retrieval and audio signal processing to analyze and represent the harmonic content of a musical piece or passage.

Specifically, chroma features aim to represent the distribution of musical notes and chords within a short-time window of audio, usually on the order of tens to hundreds of milliseconds. By analyzing the frequency content of the audio signal within this time window and mapping it onto the chromatic scale, the chroma features can capture the underlying harmonic structure of the music.

One way to calculate chroma features is to use a technique called the short-time Fourier transform (STFT), which divides the audio signal into short overlapping windows and computes the Fourier transform for each window. By applying a filterbank that emphasizes the frequencies corresponding to the 12 pitch classes, the resulting frequency spectrum can be mapped onto the chromatic scale to obtain the chroma features.

Chroma features can be used for a variety of tasks in music analysis and processing, such as chord recognition, key detection, and melody extraction. They have proven to be effective in capturing the harmonic content of music and have become a widely used tool in the field of music information retrieval.
Harmonic content refers to the set of frequencies present in a sound signal that are related to each other by integer multiples of a fundamental frequency.

In the context of music, the harmonic content of a musical passage refers to the set of musical notes and chords that make up the underlying tonal structure of the piece. These notes and chords are related to each other harmonically, meaning that they follow the rules of harmonic progression and create a sense of tonality or key.

For example, in a C major chord, the notes C, E, and G are related harmonically because they are all integer multiples of the fundamental frequency of C. Similarly, in a melody that follows a particular key, the notes used in the melody will be related harmonically to the key center and to each other, creating a sense of tonality and harmony.

Analyzing the harmonic content of a musical passage can provide insight into its structure and meaning, and can be useful for tasks such as chord recognition, key detection, and harmonic analysis. #### Chord
A chord is a group of three or more notes played simultaneously to create a harmonious sound. Chords are an important part of Western music, and are used in a wide range of genres from classical music to pop and rock.
Chords are usually constructed from a root note and a combination of other notes, which can include third, fifth, seventh, and other notes that can be added for different harmonic effects. The specific combination of notes used to construct a chord will determine its quality, such as major, minor, diminished, augmented, and others.
For example, a C major chord is composed of the notes C, E, and G played together, while a C minor chord is composed of the notes C, E-flat, and G played together. The difference in the quality of the chords is determined by the use of the third note, which is either a major third or a minor third interval above the root note.
Chords can be used in a variety of ways in music, such as to create harmony, provide accompaniment for a melody, or to create tension and release within a musical passage. They are an essential tool for composers, arrangers, and performers to create and convey the musical emotions and ideas they wish to express.
#### Differences b/w chords
The differences between chord types are determined by the specific combination of notes used to construct the chord. The main factors that determine the quality of a chord are the intervals between the notes, the root note, and the presence or absence of certain additional notes.
Here are some of the most common chord types and their defining characteristics:
Major chords - composed of a root note, a major third interval, and a perfect fifth interval above the root. They have a bright, happy sound and are commonly used in pop, rock, and classical music.
Minor chords - composed of a root note, a minor third interval, and a perfect fifth interval above the root. They have a darker, sadder sound and are commonly used in blues, jazz, and rock music.
Dominant seventh chords - composed of a root note, a major third interval, a perfect fifth interval, and a minor seventh interval above the root. They have a strong, tense sound and are commonly used in blues, jazz, and rock music.
Diminished chords - composed of a root note, a minor third interval, and a diminished fifth interval (which is also known as a tritone) above the root. They have a dissonant, unstable sound and are commonly used in classical and jazz music.
Augmented chords - composed of a root note, a major third interval, and an augmented fifth interval above the root. They have a tense, unstable sound and are commonly used in jazz and classical music.
Other chord types, such as suspended chords, add chords, and extended chords, involve the addition of other notes beyond the basic triad of root, third, and fifth intervals. These additional notes can add complexity and richness to the sound of the chord, and are commonly used in jazz and popular music.
#### Interval
In music theory, an interval is the distance between two pitches or notes. Intervals are measured by the number of letter names or staff positions between the two notes, and are described by their quality (major, minor, perfect, augmented, or diminished).
For example, the distance between the notes C and E is a third, because there are two letter names (C and D) and two staff positions (the lines or spaces on the music staff) between the two notes. The quality of this interval depends on the specific notes used and the distance between them. In this case, the interval is a major third because it spans four half-steps (two whole steps) between the notes.

Intervals are important in music theory because they are used to build scales, chords, and melodies. The specific intervals used can create different emotional effects and moods in the music, and can be used to create tension and release or to convey a sense of stability and resolution. By understanding intervals, musicians can better analyze, understand, and create music.

```python
import librosa
import librosa.display
import matplotlib
import matplotlib.pyplot as plt

# Load audio file
audio_file_path = './download1.wav'
y, sr = librosa.load(audio_file_path)

# Compute chroma features
chroma = librosa.feature.chroma_stft(y=y, sr=sr)

# Visualize chroma features
plt.figure(figsize=(5, 4))
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title('Chroma Features')
plt.tight_layout()
plt.show()
```
